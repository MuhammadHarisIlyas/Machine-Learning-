Data visualization

```python
# types of data
# Numerical data
# Categorical data
# import the libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# 2D line plot
price = [48000,54000,57000,49000,47000,45000]
year = [2015,2016,2017,2018,2019,2020]

plt.plot(year,price)
```

```python
batsman = pd.read_csv("sharma-kohli.csv")
batsman
```

```python
plt.plot(batsman['index'],batsman['V Kohli'])
```

```python
plt.plot(batsman['index'],batsman['V Kohli'])
plt.plot(batsman['index'],batsman['RG Sharma'])
```

```python
plt.plot(batsman['index'],batsman['V Kohli'])
plt.plot(batsman['index'],batsman['RG Sharma'])

plt.title('Rohit Sharma Vs Virat Kohli Career Comparison')
plt.xlabel('Season')
plt.ylabel('Runs Score')
```

```python
# to change the color of lines in graph
plt.plot(batsman['index'],batsman['V Kohli'],color = 'green')
plt.plot(batsman['index'],batsman['RG Sharma'], color = 'black')

plt.title('Rohit Sharma Vs Virat Kohli Career Comparison')
plt.xlabel('Season')
plt.ylabel('Runs Score')
```

```python
plt.plot(batsman['index'],batsman['V Kohli'],linestyle='dashed')
plt.plot(batsman['index'],batsman['RG Sharma'],linestyle='dashed')

plt.title('Rohit Sharma Vs Virat Kohli Career Comparison')
plt.xlabel('Season')
plt.ylabel('Runs Score')
```

```python
plt.plot(batsman['index'],batsman['V Kohli'],color = 'green',label='Virat')
plt.plot(batsman['index'],batsman['RG Sharma'], color = 'black',label='Rohit')

plt.title('Rohit Sharma Vs Virat Kohli Career Comparison')
plt.xlabel('Season')
plt.ylabel('Runs Score')

plt.legend()
```

```python
# limiting axes
price = [48000,54000,57000,49000,47000,45000,4500000]
year = [2015,2016,2017,2018,2019,2020,2021]

plt.plot(year,price)
plt.ylim(0,75000)
plt.xlim(2017,2019)
```

```python
# grids

plt.plot(batsman['index'],batsman['V Kohli'],linestyle='dashed')
plt.plot(batsman['index'],batsman['RG Sharma'],linestyle='dashed')

plt.title('Rohit Sharma Vs Virat Kohli Career Comparison')
plt.xlabel('Season')
plt.ylabel('Runs Score')

plt.grid()
```

```python
# show

plt.plot(batsman['index'],batsman['V Kohli'],linestyle='dashed')
plt.plot(batsman['index'],batsman['RG Sharma'],linestyle='dashed')

plt.title('Rohit Sharma Vs Virat Kohli Career Comparison')
plt.xlabel('Season')
plt.ylabel('Runs Score')

plt.grid()

plt.show()
```

```python
# plt.scatter simple function
x = np.linspace(-10,10,50)

y = 10*x + 3 + np.random.randint(0,300,50)
y
```

```python
plt.scatter(x,y)
```

```python
# plt.scattter on pandas data

batter = pd.read_csv('batter.csv')
batter = batter.head(50)
batter
```

```python
plt.scatter(batter['avg'],batter['strike_rate'])
plt.title('Avg and SR analysis of top 50 Batsman')
plt.xlabel('Average')
plt.ylabel('SR')
```

```python
# Size
# Slower technique
tips = sns.load_dataset('tips')

plt.scatter(tips['total_bill'],tips['tip'],s = tips['size']*20)
```

```python
# scatter using plt.plot
# faster technique

plt.plot(tips['total_bill'],tips['tip'],'o')
```

Bar chart

```python
# Simple bar chart
children = ['10','20','60','40','50']
colors = ['red','green','blue','yellow','pink']

plt.bar(colors,children, color='black')
```

```python
# horizontal bar chart
plt.barh(colors,children, color='black')
```

```python
# colors and labels
df = pd.read_csv('batsman_season_record.csv')
df
```

```python
# multiple bar chart
plt.bar(np.arange(df.shape[0]) - 0.2,df['2015'],width=.2)
plt.bar(np.arange(df.shape[0]),df['2016'],width=.2)
plt.bar(np.arange(df.shape[0]) + 0.2,df['2017'],width=.2)

#xticks

plt.xticks(np.arange(df.shape[0]),df['batsman'])
plt.show()
```

```python
plt.bar(df['batsman'],df['2017'],label='2017')
plt.bar(df['batsman'],df['2016'],bottom=df['2017'],label='2016')
plt.bar(df['batsman'],df['2015'],bottom=(df['2016'] + df['2017']),label='2015')

plt.legend()
plt.show()
```

# Pie chart

```python
# Simple data
data = [23,50,100,29,65]
subjects = ['eng','maths','science','sst','urdu']
plt.pie(data,labels=subjects)
plt.show()
```

```python
# dataset
gayle = pd.read_csv('gayle-175.csv')
gayle
```

```python
plt.pie(gayle['batsman_runs'],labels=gayle['batsman'], autopct='%0.1f%%')
plt.show()
```

```python
# percentage and color
plt.pie(gayle['batsman_runs'],labels=gayle['batsman'], autopct='%0.1f%%', colors=['green','blue','yellow','red','pink','brown'])
plt.show()
```

```python
# explode shadow
plt.pie(gayle['batsman_runs'],labels=gayle['batsman'], autopct='%0.1f%%', explode=[.1,0,0,0,0,0])
plt.show()
```

```python
plt.pie(gayle['batsman_runs'],labels=gayle['batsman'], autopct='%0.1f%%', shadow=True)
plt.show()
```
